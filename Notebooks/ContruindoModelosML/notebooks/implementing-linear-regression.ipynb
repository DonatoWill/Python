```python
import pandas as pd
```

```python
auto_data = pd.read_csv('../data/imports-85.data', sep=r'\s*,\s*', engine='python')
# The data is separate by ,
```

```python
auto_data
```

```python
import numpy as np

auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe()
```

```python
# We nedd convert this column to be of type float for use in the model
# errors= coerce ignore errors and force the convertion
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
```

```python
auto_data['price'].describe()
```

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['num-of-cylinders'].describe()
```

```python
cylinders_dict = {'two': 2,
                  'three': 3,
                  'four': 4,
                  'five': 5,
                  'six':6,
                  'eight': 8,
                  'twelve': 12,}

auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)

auto_data.head()
```

```python
# Using get dummies to convert categorical to one-hot representation

auto_data = pd.get_dummies(auto_data, columns=['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style',
                                               'drive-wheels', 'engine-location', 'engine-type', 'fuel-system'])
auto_data.head()
```

```python
# Now all columns are numerical
auto_data.info()
```

```python
auto_data = auto_data.dropna()
auto_data.info()
```

```python
# Now we have a clean data
from sklearn.model_selection import train_test_split

# Drop the target (label) column
X = auto_data.drop('price', axis=1)

# Take the labels (price) to train
Y = auto_data['price']

# Spliting into 80% for training set and 20% for testing set
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
# copy_X: this means that X variables that we've passed in will be copied over, will not be overwritten
# fit_intercept by default is set to True, indicating that intercepts ought to be used. 
# The data is note centered around zero
# normalize: indicates that we don't want use normalization, that is similar to standardization 
```

```python
# Check how good the model is, this is the R² value of prediction
linear_model.score(X_train, Y_train)
```

```python
linear_model.coef_
```

```python
predictors = X_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()

print(coef)
# Some features are negatively correlated, it means that cars with these are cheaper
# Some features are colse to zero wich means that features have very little effect on vehicle price
# The highest coefficient values are factor wich contribute to a high vehicle price
```

```python
y_predict = linear_model.predict(x_test)
```

```python
# Here we plot the predicted values and the actual values to compare how close they are
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')

plt.legend()
plt.show()
```

```python
r_square = linear_model.score(x_test, y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict, y_test)
linear_model_mse
```

```python
import math

math.sqrt(linear_model_mse)
# It means on average the price predicted by our model is around + or - 5.100 away from thr actual 
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso

# alpha is the regularization parameter to prevent overfitting of the model
# Increasing the value of alpha has forced more significant feature to drive our prediction
lasso_model = Lasso(alpha=4, normalize=True) 
lasso_model.fit(X_train, Y_train)
```

```python
lasso_model.score(X_train, Y_train)
```

```python
coef = pd.Series(lasso_model.coef_, predictors).sort_values()
print(coef)
```

```python
y_predict = lasso_model.predict(x_test)
```

```python
pylab.rcParams['figure.figsize'] = (15, 6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')

plt.legend()
plt.show()
# Here we can see the fit is much better
```

```python
r_square = lasso_model.score(x_test,y_test)
r_square
```

```python
lasso_model_mse = mean_squared_error(y_predict, y_test)
math.sqrt(lasso_model_mse)
```
